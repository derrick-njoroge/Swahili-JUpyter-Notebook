```python
import re
from nltk.stem import PorterStemmer
import pandas as pd

def preprocess_text(tweet_df):
    tokenized_tweet = tweet_df['content'].apply(lambda x: x.split())

    stemmer = PorterStemmer()

    stemmed_tweet = tokenized_tweet.apply(lambda x: [stemmer.stem(token) for token in x])

    swahili_stop_words = ['ya', 'na', 'kwa', 'ni', 'wa', 'wake', 'ya', 'yake', 'yangu', 'yao', 'yeye', 'yote', 'yoyote', 'yule', 'za', 'zaidi', 'zake', 'zangu', 'za', 'zipi', 'zote']
    
    filtered_tweet = stemmed_tweet.apply(lambda x: [word for word in x if word not in swahili_stop_words])
    tweet_df['filtered_tokens'] = filtered_tweet
    
    tweet_df['processed_text1'] = tweet_df['filtered_tokens'].apply(lambda x: ' '.join(x))
    
    return tweet_df

    
```

```python
tweet_df = pd.read_csv('Walibora.csv')

preprocessed_tweet_df = preprocess_text(tweet_df)

preprocessed_tweet_df.to_csv('processed_tweets1.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import random
import re
import string
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


preprocessed_tweet_df = pd.read_csv('processed_tweets1.csv')
```

Training Model 

```python
X_train, X_test, y_train, y_test = train_test_split(preprocessed_tweet_df['processed_text1'], preprocessed_tweet_df['content'], test_size=0.3, random_state=42)

vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

nb_model = MultinomialNB()
nb_model.fit(X_train_vec, y_train)

y_pred = nb_model.predict(X_test_vec)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
from sklearn.metrics import classification_report

y_pred = nb_model.predict(X_test_vec)
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import re

preprocessed_tweet_df = pd.read_csv("processed_tweets1.csv")

vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(preprocessed_tweet_df['processed_text1'])
nb_model = MultinomialNB()
nb_model.fit(X_train_vec, preprocessed_tweet_df['content'])

def classify_new_text(new_text):
    new_text = re.sub('[^a-zA-Z0-9\s]+', '', new_text) 
    new_text = new_text.lower()  
    new_text = new_text.strip()  
    new_text = re.sub('\s+', ' ', new_text)  
    new_text_vec = vectorizer.transform([new_text])
    predicted_class = nb_model.predict(new_text_vec)[0]
    return predicted_class
```
